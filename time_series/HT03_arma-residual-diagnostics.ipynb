### Estimating the 'Best' Model

```python
import statsmodels.tsa.api as tsa
```

```python
# Estimating our model
mod = tsa.SARIMAX(term, order=(1,0,1), trend="c")
res = mod.fit()
```

### Plot the residuals and check for serial correlation

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Doing the plot
sns.set_style("darkgrid")
plt.rc("figure", figsize=(16,6))
plt.rc("hist", bins=41)
fig = res.plot_diagnostics()
```

The main thing we are looking at when analysing residuals from a time series model is to check if we have modeled all the time series dependencies in the data. That is, if the fitted residuals exhibit serial correlation or not.

Do these errors seem to have serial correlation? It doesn't look like it. But we need further analysis. Note that the error terms are obviously heteroskedastic. In the beginning of the 1980s, at a time of high inflation, we observed the largest error terms. 

It might help to compare these errors with those of a random-walk model. Remember that a random-walk model is defined as:

$ y_t = y_{t-1} + \epsilon_t $

Therefore, the residuals of a random-walk model can be written as:

$ \hat{\epsilon}_t = y_t - y_{t-1} $

With this in mind, lets compare both residuals:

```python
df = pd.DataFrame([term.diff(), res.resid], index=["RW", "ARMA(1,1)"]).T
sns.set_style("darkgrid")
df.plot(figsize=(16,6))
```

The residuals from the random-walk model appear to have more dynamics in them than the ones from the ARMA(1,1) model. But to know if we are indeed modelling all the time series dependencies, let's take a look at some formal tests for residual serial correlation.

### Ljung Box Test 

From page 253 of Kevin's notes we have:

```python
from statsmodels.stats.diagnostic import acorr_ljungbox
import numpy as np
```

```python
# Saving the residuals from our model
resid = res.resid
# Doing the Ljung Box Test
stat, pval = acorr_ljungbox(resid, lags=12)
```

```python
# Plotting the p-values of our Ljung Box Test
plt.rc("figure", figsize=(16,6))
plt.bar(np.arange(1, 13), pval)
plt.plot([0,13],[0.05,0.05], linestyle=":", color="black")
ax = plt.gca()
ax.set_title("Ljung-Box Q P-value")
ax.set_xlabel("Lags Included")
lim = ax.set_xlim(0.5,12.5)
```

Since the null in the Ljung Box Test is that there is no serial correlation in the residuals up to the lag being tested, the p-values above contribute to the hypothesis that is no serial correlation in the residuals of our ARMA(1,1) model. 

For comparison, let's see how the results of this test for the residuals of the random-walk model look like:

```python
# Saving the residuals from RW model
resid_rw = term.diff().iloc[1:]
# Doing the Ljung Box Test
stat_rw, pval_rw = acorr_ljungbox(resid_rw, lags=12)
```

```python
# Plotting the p-values of our Ljung Box Test
plt.rc("figure", figsize=(16,6))
plt.bar(np.arange(1, 13), pval_rw)
plt.plot([0,13],[0.05,0.05], linestyle=":", color="black")
ax = plt.gca()
ax.set_title("Ljung-Box Q P-value")
ax.set_xlabel("Lags Included")
lim = ax.set_xlim(0.5,12.5)
```

```python
pval_rw
```

As we can see above, there is strong evidence of serial correlation in the residuals of the random-walk model. 

This is all very interesting. But because the Ljung Box Test assumes homoskedasticity, it is not very useful when dealing with financial data. Now let's proceed to the heteroskedasticy robust LM Test

### LM Test for Serial Correlation 

From page 253 of Kevin's notes we have:

```python
from statsmodels.stats.diagnostic import acorr_lm
```

```python
# Doing the LM Test for our model
pval = {}
for i in range(1, 13):
    stat, pv, _, _ = acorr_lm(resid, i, None)
    pval[i] = pv
pval = pd.Series(pval)    
```

```python
# Plotting the results
plt.rc("figure", figsize=(16,6))
colors = sns.color_palette()
plt.bar(np.arange(1, 13), pval, color=colors[1])
plt.plot([0,13],[0.05,0.05], linestyle=":", color="black")
ax = plt.gca()
ax.set_title("LM Test of Serial Correaltion P-value")
ax.set_xlabel("Lags Included")
lim = ax.set_xlim(0.5,12.5)    
```

We can see that by using a heteroskedasticity robust version of a test for serial correlation, we have an even stronger evidence that the residuals of our ARMA(1,1) do not present serial correlation.

Let's conduce the LM test for serial correlation using the residuals of the random-walk model for comparison:

```python
# Doing the LM Test for RW model
pval_rw = {}
for i in range(1, 13):
    stat, pv, _, _ = acorr_lm(resid_rw, i, None)
    pval[i] = pv
pval_rw = pd.Series(pval) 
```

```python
# Plotting the results
plt.rc("figure", figsize=(16,6))
colors = sns.color_palette()
plt.bar(np.arange(1, 13), pval_rw, color=colors[1])
plt.plot([0,13],[0.05,0.05], linestyle=":", color="black")
ax = plt.gca()
ax.set_title("LM Test of Serial Correaltion P-value")
ax.set_xlabel("Lags Included")
lim = ax.set_xlim(0.5,12.5)   
```

```python
pval_rw
```

Again, there is strong evidence of serial correlation in the residuals of the random-walk model. This is because the random walk model does not fully capture the temporal dependence of the data. Overall, the results from this section makes us feel very confident that our ARMA(1,1) does a good job in capturing this temporal dependence observed in the data.
